# Data Cleaning

## Removing the Outliers
1. Z_Score
2. IQR range

```python
### Outliers removed by  IQR method
```

```python
for colname,colvals in df.iteritems():
    if df[colname].dtype!=object and colname!='Profit':
        Q1=df[colname].quantile(q=0.25)
        Q3=df[colname].quantile(q=0.75)
        IQR=Q3-Q1
        low=Q1-(1.5*IQR)
        up=Q3+(1.5*IQR)
        df.loc[(df[colname]<low) | (df[colname]>up) ,colname]=df[colname].median()
    else:
        continue
```

# Inclusion of new Features :
### 1.Net Sales= Gross Sales - Discount
### 2.cost

```python
def net_sales(retail_data):
    net_sales=retail_data['Sales'] * (1-retail_data['Discount'])
    return net_sales
retail_data['Net Sales']= retail_data.apply(net_sales, axis=1)
```

```python
retail_data.head()
```

```python
def Cost(retail_data):
    Cost=retail_data['Net Sales'] - (retail_data['Profit'])
    return Cost
retail_data['Cost']= retail_data.apply(Cost, axis=1)
```
